```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://en.wikipedia.org/wiki/Krasnoyarsk_Pillars'
r_wiki = requests.get(url=url)
```

```python
r_wiki
```

```python
r_wiki.text
```

```python

soup_wiki = BeautifulSoup(r_wiki.text, 'html.parser')
soup_wiki
```

Вытаскиваем все заголовки h2 со страницы — изучаем структуру

```python
pars_wiki = soup_wiki.find_all('h2')
pars_wiki
```

Тут все не очень интересно. Поэтому выведем весь текст

```python
pars_wiki = soup_wiki.find_all('p')
pars_wiki
```

Теперь причешим содержимое и выведем тексты.

```python
texts_wiki = [text.get_text() for text in pars_wiki]
texts_wiki
```
